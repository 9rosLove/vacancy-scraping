```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('vacancy/vacancies.csv')

technologies_count = df['technologies'].str.split(',').explode().str.strip().value_counts()
technologies_count["DRF"] = technologies_count["Django REST Framework"] + technologies_count["DRF"]
technologies_count.drop("Django REST Framework", inplace=True)
technologies_count.sort_values(ascending=False, inplace=True)
top_technologies = technologies_count.head(25)[::-1]
top_technologies.plot(kind='barh', figsize=(15, 9), color='darkblue', width=0.8)

plt.title('Top 25 mentioned technologies')
plt.xlabel('Frequency')
plt.ylabel('Technologies')
plt.show()
```
